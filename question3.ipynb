```python
import pandas as pd
import numpy as np
import random

seriesA = pd.Series(np.random.randint(1,5,100))
seriesB = pd.Series(np.random.randint(1,4, 100))
seriesC = pd.Series(np.random.randint(10000,30001, 100))
print(seriesA)
print(seriesB)
print(seriesC)
```

```python
JoinSeries = pd.DataFrame({'SA':seriesA, 'SB':seriesB, 'SC':seriesC})
JoinSeries.head()
```

```python
JoinSeries.rename(columns={'SA':'bedrs', 'SB':'bathrs', 'SC':'price_sqr_meter' }).head()
```

```python
JoinSeries.columns = ['bedrs', 'bathrs', 'price_sqr_meter']
JoinSeries.head()
```

```python
BigColumn = pd.DataFrame()
BigColumn['bigcolumn'] = JoinSeries['bedrs'].append(JoinSeries['bathrs']).append(JoinSeries['price_sqr_meter']).reset_index(drop=True)
BigColumn
```
