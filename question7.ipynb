# Question 7
Create a vertical bar chart comparing the number of marriages and divorces per
capita in the U.S. between 1900, 1950, and 2000.
Don't forget to label your axes!

```python
import pandas as pd
import matplotlib.pyplot as mpl
%matplotlib inline
 
us_mrg_div = pd.read_csv('us-marriages-divorces-1867-2014.csv')
 
us_mrg_div.set_index('Year', inplace = True)
 
plot_chart = us_mrg_div.loc[[1900,1950,2000],['Marriages_per_1000','Divorces_per_1000']]
plot_chart.plot.bar(title="Comparison of 1900, 1950, 2000", figsize=(8,4))
```
